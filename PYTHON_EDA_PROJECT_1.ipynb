# Gender

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount' ,data=sales_gen)
```

```python
From above graph we can see that purchasing power of female are greater than men,  
Also we can see that most of the buyers are females
```

# Age

```python
sns.countplot(data=df,x= 'Age Group',hue='Gender')  ##hue simply divide the data on the man or woman basis
```

```python
ax=sns.countplot(data=df,x= 'Age Group',hue='Gender') 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount' ,data=sales_age)
```

```python
From above graphs we can see that most of the buyers are of the age group between 26-35 yrs female
```

# State

```python
df.columns
```

```python
#Totalno.of.orders from top 8 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y=  'Orders',data=sales_state)
```

```python
#Total amount/sales from top 8 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y=  'Amount',data=sales_state)
```

```python
From above graphs we can see that  most of the order are From uttar pradesh , maharashtra And karnataka ,
Respectively tatal sale/amount Is From uttar pradesh , maharashtra And karnataka
```

# Marital status

```python
ax=sns.countplot(data=df,x= 'Marital_Status' ) 
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y=  'Amount',hue='Gender',data=sales_state)
```

```python
From above graph  we can see that most of the buyers are married (women) and they have high purchasing power.
```

# Occupation

```python
ax=sns.countplot(data=df,x= 'Occupation' ) 
sns.set(rc={'figure.figsize':(20,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y=  'Amount',data=sales_state)
```

```python
From above graph  we can see that most of the buyers are working In IT ,Aviation And heathcare sector
```

# Product Category

```python
ax=sns.countplot(data=df,x= 'Product_Category' ) 
sns.set(rc={'figure.figsize':(20,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,4)})
sns.barplot(x='Product_Category', y=  'Amount',data=sales_state)
```

```python
From above graph  we can see that most of the sold products are from food, footwear, electonics category
```
